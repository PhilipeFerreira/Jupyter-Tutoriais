## Importing the Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting

You can read much more about the **plot** function in the <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot">documentation</a>.

```python
x = [0,1,2]
y = [0,1,4]
```

```python
%matplotlib inline
fig = plt.figure() 
axes = fig.add_subplot(111) 

axes.plot(x, y)

plt.show()
```

```python
x_highres = np.linspace(0, 2, 20)
y_highres = x_highres ** 2

fig = plt.figure()
axes = fig.add_subplot(111)

axes.plot(x,y)
axes.plot(x_highres, y_highres)

plt.show()
```

### Controlling Line Style 

```python
fig = plt.figure()

axes = fig.add_subplot(111)
#axes.plot(x_highres, y_highres, "r--")
axes.plot(x_highres, y_highres, color='red', linestyle='dashed') #same as line above

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_subplot(111)
axes.plot(x_highres, y_highres, color='red', linestyle='dashed', linewidth=3, marker='o', markerfacecolor='blue', markersize=5)

plt.show()
```

### Title and Grid

```python
fig = plt.figure()

axes = fig.add_subplot(111)
axes.plot(x_highres, y_highres)
axes.set_title('$y=x^2$') # Latex Code
axes.grid()

plt.show()
```

### Axis Labels

```python
fig = plt.figure()

axes = fig.add_subplot(111)
axes.plot(x_highres, y_highres)
axes.set_title('$y=x^2$') # Latex Code
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel('y')

plt.show()
```

### Controlling the Figure Size

```python
fig = plt.figure(figsize=(12,8)) # here you set the figure size

axes = fig.add_subplot(111)
axes.plot(x_highres, y_highres)
axes.set_title('$y=x^2$') # Latex Code
axes.grid()

plt.show()
```

### Visualizing Numbers as images

```python
noise = np.random.random((128,128))
noise
```

```python
plt.imshow(noise)
plt.show()
```

### Color Bar

```python
plt.imshow(noise)
plt.colorbar()
plt.show()
```

### Color Maps

```python
plt.imshow(noise, cmap=plt.cm.gray)
plt.colorbar()
plt.show()
```

```python
plt.imshow(noise, cmap=plt.cm.Paired)
plt.colorbar()
plt.show()
```

### Multiples Plot at Same Graph

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.bar()
plt.show()
```

### Subplots

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid with height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# The 1st plot
plt.plot(x, y_sin)
plt.title('Sine')

# The 2nd plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure
plt.show()
```
